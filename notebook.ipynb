```python
import json
import os
```

```python
folder_path = "Torah Text JSON Files"

full_chap_length = {}

for filename in os.listdir(folder_path):
    name = filename.split('.')[0]
    print(name)
    file_path = os.path.join(folder_path, filename)
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    text = data['text']
    chap_length = {chap_num + 1 : len(chap) for chap_num, chap in enumerate(text)}
    full_chap_length[name] = chap_length
```

```python
full_chap_length['Genesis'] = full_chap_length['genesis']
full_chap_length['Numbers'] = full_chap_length['numbers']
full_chap_length['Exodus'] = full_chap_length['exodus']
full_chap_length['Leviticus'] = full_chap_length['leviticus']
full_chap_length['Deuteronomy'] = full_chap_length['deuteronomy']
```

```python
parsha_file = "Parsha Aliyah.json"
with open(parsha_file, 'r', encoding='utf-8') as filename:
    parsha_data = json.load(filename)
```

```python
def calc_length(chap_length, chapter_start, verse_start, chapter_end, verse_end):
    length = 0

    if chapter_start == chapter_end:
        length = verse_end - verse_start + 1
    else:
        length += chap_length[chapter_start] - (verse_start - 1)

        for chap in range(chapter_start + 1, chapter_end):
            length += chap_length[chap]
        length += verse_end

    return length



for parsha in parsha_data:
    length = 0
    chap_length = full_chap_length[parsha['Book']]
    cstart = int(parsha['chapter start'])
    vstart = int(parsha['verse start'])
    cend = int(parsha['chapter end'])
    vend = int(parsha['verse end'])

    length = calc_length(chap_length, cstart, vstart, cend, vend)
    parsha['length'] = length
```

```python
data = {parsha['Name'] : parsha['length'] for parsha in parsha_data}
```

```python
with open('parsha length.json', 'w', encoding='utf-8') as json_output:
    json.dump(data, json_output, ensure_ascii=False, indent=2)
```

```python
def parse_verses_info(verse_info):
    # Splitting the verses string into parts
    parts = verse_info["verses"].split('-')

    # Extracting chapter start, verse start, chapter end, and verse end values
    chapter_start, verse_start = map(int, parts[0].split(':'))
    chapter_end, verse_end = map(int, parts[1].split(':'))

    # Adding keys to the input dictionary
    verse_info["chapter_start"] = chapter_start
    verse_info["verse_start"] = verse_start
    verse_info["chapter_end"] = chapter_end
    verse_info["verse_end"] = verse_end

    return verse_info
```

```python
for parsha in parsha_data:
    parsha['Aliyot'] = [parse_verses_info(aliyah) for aliyah in parsha['Aliyot']]
        
```

```python
for parsha in parsha_data:
    chap_length = full_chap_length[parsha['Book']]
    for aliyah in parsha['Aliyot']:
        length = 0
        cstart = int(aliyah['chapter_start'])
        vstart = int(aliyah['verse_start'])
        cend = int(aliyah['chapter_end'])
        vend = int(aliyah['verse_end'])
        length = calc_length(chap_length=chap_length, chapter_end=cend, chapter_start=cstart, verse_end=vend, verse_start=vstart)
        aliyah['length'] = length
```

```python
aliyah_lengths = [
    {
    'name' : parsha['Name'],
    'length' : parsha['length'],
    'book' : parsha['Book'],
    'aliyah_lengths' : [
        aliyah['length']

        for aliyah in parsha['Aliyot']
    ]
    
    }
    for parsha in parsha_data]
```
